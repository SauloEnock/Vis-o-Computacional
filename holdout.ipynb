```python
from ultralytics.utils import SETTINGS

SETTINGS["raytune"] = False  # Disable ray.tune integration
```

```python
from hyperopt import fmin, tpe, hp, Trials

def loss(params):
    # Training.
    imgsz, epochs, batch, optimizer, lr0 = params

    default_args2 = {
        'imgsz': imgsz,
        'epochs': epochs,
        'batch': batch,
        'optimizer': optimizer,
        'lr0': lr0,
        'amp': False,
    }
    
    results = model.train(
       data='/kaggle/working/dataHoldOut.yaml',
       name=model_name, **default_args2)
    
    
    metrics = model.val(
       data='/kaggle/working/dataHoldOut.yaml', **default_args2)  # assumes `model` has been loaded
    
    return (1 - metrics.box.map)  # mAP50-95
```

```python
space = [
    # hp.loguniformuniform('lrh',0,0.00001),
     hp.choice('imgsz', [128, 256]),
     hp.choice('epochs', [25, 20, 15, 10]),
     hp.choice('batch', [16, 12, 8, 4]), 
     hp.choice('optimizer', ['SGD', 'Adam', 'AdamW', 'NAdam', 'RAdam', 'RMSProp']), 
     hp.uniform('lr0', 0.001, 0.01), 
]
```

```python
best = fmin(loss, space, algo=tpe.suggest, max_evals=20)
```

```python
imgsz = [128, 256]
epochs = [25, 20, 15, 10]
batch = [16, 12, 8, 4]
optimizer = ['SGD', 'Adam', 'AdamW', 'NAdam', 'RAdam', 'RMSProp']

print("Melhores parâmetros encontrados:")
print("imgsz: ", imgsz[best['imgsz']])
print("epochs: ", epochs[best['epochs']])
print("batch: ", batch[best['batch']])
print("optimizer: ", optimizer[best['optimizer']])
print("Learning rate: ", best['lr0'])
```
